```python
import pandas as pd
import re
import json
import requests
import numpy as np
from os import walk, getcwd
```

```python
# folder path
dir_path = getcwd() + '/raw_Data'

# list to store files name
raw_data = []
for (dir_path, dir_names, file_names) in walk(dir_path):
    raw_data.extend(file_names)


for file_name in raw_data:
    data_path = f'raw_Data/{file_name}' 
    print(data_path)
    df = pd.read_csv(data_path)
    df['images'] = 'NaN'
    for i in range(df.shape[0]):
        print(data_path, end=' ')
        print(i)
        url = df.loc[i]['link']
        id_ = re.search(r"-(\d+)/", url).group(1)
        product_url = f"https://api.digikala.com/v1/product/{id_}/"
        data = requests.get(product_url).json()
        if 'seo' in data['data']:
            df.iat[i, 5] = data['data']['seo']['markup_schema'][0]['image']
        else:
            df.drop(axis=0, index=i)
    print(df.head(5))
    print()
    df.to_csv(f'clean_Data/{file_name}')
```
